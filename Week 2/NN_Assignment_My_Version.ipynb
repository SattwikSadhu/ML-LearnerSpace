```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,regularizers
```

```python
!unzip homer_bart.zip -d homer_bart
```

```python
ds_preprocessed = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/homer_bart",
    labels="inferred",
    label_mode="binary",
    # class_names=['0', '1', '2', '3', ...]
    image_size=(64,64),
    seed=0,
    validation_split=0.1,
    subset="both")
```

```python
ds_train=ds_preprocessed[0] #training dataset
ds_test=ds_preprocessed[1]  #test dataset
```

```python
model = keras.Sequential()
model.add(keras.Input(shape=(64,64,3)))
model.add(layers.Flatten())
model.add(layers.Rescaling(1./255))
#model.add(layers.Dense(32*32*3, activation="relu"))
#model.add(layers.Dense(16*16*3, activation="relu",kernel_regularizer=regularizers.l2(.01)))
#model.add(layers.Dense(8*8*3, activation="relu"))
#model.add(layers.Dense(1000, activation="relu"))
model.add(layers.Dense(100, activation="relu"))
model.add(layers.Dense(10, activation="relu"))
model.add(layers.Dense(1,activation="sigmoid"))
```

```python
print(model.summary())
```

```python
model.compile(
    loss=keras.losses.BinaryCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"])
```

```python
model.fit(ds_train,epochs=50,verbose=2,validation_data=ds_test)
```

```python
loss,accuracy=model.evaluate(ds_test, verbose=2)
print('Test Accuracy = ',accuracy*100,'%')  #Although test accuracy is good for this particular train-test split, it's not consistent for other train-test splits obtained by changing the 'seed' value even for the same model.
```
